```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing stock trend data.
stockdf = pd.read_csv('hc_df_preprocessed.csv')
stockdf.head()
```

```python
#importing covid cases data.
coviddf = pd.read_csv('covid_df_preprocessed.csv')
coviddf.head()
```

```python
#changing columns name so we can merge the data.
stockdf.rename(columns={"Date": "date"}, inplace=True)
```

```python
#Merging the data.
data = pd.merge(stockdf, coviddf, on='date')
```
